Resizing the image using Pillow library

```python
from PIL import Image
```

```python
img = Image.open('dog.jpg')
```

```python
img_resized =img.resize((200, 200))
```

```python
img_resized.save('dog_resized.jpg')
```

```python
# Displaying the image from numpy array

img_res = mpimg.imread('/content/dog_resized.jpg')
img_res_plot = plt.imshow(img_res)
```

**Convert Rgb image to Grayscale Image using OpenCV**

```python
# import OpenCV library

import cv2
```

```python
img_cv = cv2.imread('/content/dog.jpg')
```

```python
type(img_cv)
```

```python
img_cv.shape
```

```python
grayscale_image = cv2.cvtColor(img_cv, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

cv2.imshow() will display the image but this will be allowed in google collab

```python
from google.colab.patches import cv2_imshow
```

```python
# display the image

cv2_imshow(grayscale_image)
```

```python
# saving the grayscale image

cv2.imwrite('dog_grayscale_image.png' , grayscale_image)
```
