# **Train Test Split**

```python
X_train, X_test, Y_train, Y_test = train_test_split(x , y , test_size = 0.2 , stratify= y , random_state= 2)
```

```python
print(x.shape ,X_train.shape, X_test.shape)
```

**Comparing the performance of the model**



```python
# list of models
models = [LogisticRegression(max_iter=1000), svm.SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compair_model_train_test():
  for model in models:
    model.fit(X_train, Y_train)
    test_data_prediction = model.predict(X_test)
    accuracy = accuracy_score(Y_test, test_data_prediction)
    print('Accuracy score of the ', model, ' = ', accuracy)
```

```python
compair_model_train_test()
```

#  **Cross Validation**

Logisctic Resgression

```python
cv_score = cross_val_score(LogisticRegression(max_iter=1000), x, y, cv=5)

print(cv_score)

mean_accuracy = sum(cv_score)/len(cv_score)

mean_accuracy = mean_accuracy*100

mean_accuracy = round(mean_accuracy, 2)

print(mean_accuracy)
```

Support Vector Classifier

```python
cv_score_svm = cross_val_score(svm.SVC(kernel='linear'), x, y, cv=5)

print(cv_score_svm)

mean_accuracy_svm = sum(cv_score_svm)/len(cv_score_svm)

mean_accuracy_svm = mean_accuracy_svm*100

mean_accuracy_svm = round(mean_accuracy_svm, 2)

print(mean_accuracy_svm)
```

Creating the function  to compair the models

```python
# list of models
models = [LogisticRegression(max_iter=1000), svm.SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compair_model_cross_validation():

  for model in models:

    cv_score = cross_val_score(model, x, y, cv=5)

    mean_accuracy = sum(cv_score)/len(cv_score)

    mean_accuracy = mean_accuracy*100

    mean_accuracy = round(mean_accuracy, 2)

    print('Cross Validation accuracies for the' ,model ,"=" , cv_score)
    print("Accuracy % of the " , model , mean_accuracy)
    print('------------------------------------------------------')
```

```python
compair_model_cross_validation()
```
