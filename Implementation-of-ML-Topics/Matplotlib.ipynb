```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
```

```python
#adding x axis and y axis lables

plt.plot(x,y)
plt.xlabel('angles')
plt.ylabel('sin value')
plt.title('sin wave')
plt.show()
```

```python
x= np.linspace(-10 ,10 , 20)
y = x**2

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y , 'r*')
plt.show()
```

```python
x = np.linspace(-5 , 5 ,100)
plt.plot(x , np.sin(x) , 'b-')
plt.plot(x , np.cos(x) , 'r--')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langagues = ['English' , 'Hindi' , 'Marathi' , 'Tamil' ,'German']
people = [100 , 120 , 145 , 130 , 90]
ax.bar(langagues , people)
plt.xlabel('LANGAGUES')
plt.ylabel('PEOPLE')
plt.show()
```

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
langagues = ['English' , 'Hindi' , 'Marathi' , 'Tamil' ,'German']
people = [100 , 120 , 145 , 130 , 90]
ax1.pie(people , labels = langagues, autopct = '%1.2f%%')
plt.show()
```

```python
x = np.linspace(0,10,80)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax2 = fig2.add_axes([0,0,1,1])
ax2.scatter(x,y , color = 'g')
ax2.scatter(x,z , color = 'r')
plt.show()
```

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 40* np.random.random(10000)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z, c = z , cmap = 'Greens')
plt.show()
```
