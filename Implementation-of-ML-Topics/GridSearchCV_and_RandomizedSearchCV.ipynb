Importing the dependencies

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

**Data collection & Processing**

```python
# loding the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loding the data to a pandas dataframe
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# adding the 'target' column to our dataset
data_frame['label'] = breast_cancer_dataset.target
```

```python
# getting some info about the dataset
data_frame.info()
```

# seprate the feature and Target

```python
X = data_frame.drop(columns = 'label', axis = 1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

# **GridserchCV**

GridserchCV is used for determining the best parameters for our model

```python
# loading the svc model
model = SVC()
```

```python
# Hyperparamaeters

parameters = {
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
# grid serch

Classifier = GridSearchCV(model, parameters, cv=5)
```

```python
# fitting the data to our model
Classifier.fit(X,Y)
```

```python
Classifier.cv_results_
```

```python
# best parameters

best_parameters = Classifier.best_params_
print(best_parameters)
```

```python
# highest accuracy

highest_accuracy = Classifier.best_score_
print(highest_accuracy)
```

```python
# loding the results to a pandas dataframe
results = pd.DataFrame(Classifier.cv_results_)
```

```python
results.head()
```

```python
grid_serch_results = results[['param_C','param_kernel','mean_test_score']]
```

```python
grid_serch_results
```

Highest accuracy = 95.2%

best paerameters = {'C': 10 , 'Kernal' = 'liner'}

# **RandomizedSerchCV**

```python
# loading the svc model
model = SVC()
```

```python
# Hyperparamaeters

parameters = {
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
# grid serch

Classifier = RandomizedSearchCV(model, parameters, cv=5)
```

```python
# fitting the data to our model
Classifier.fit(X,Y)
```

```python
Classifier.cv_results_
```

```python
# best parameters

best_parameters = Classifier.best_params_
print(best_parameters)
```

```python
# highest accuracy

highest_accuracy = Classifier.best_score_
print(highest_accuracy)
```

```python
# loding the results to a pandas dataframe
results = pd.DataFrame(Classifier.cv_results_)
```

```python
results.head()
```

```python
Randomized_serch_results = results[['param_C','param_kernel','mean_test_score']]
```
