Importing The Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
import sklearn.datasets
```

Import The California House Price Dataset

```python
house_price_datset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_datset)
```

```python
# Loding dataset to a pandas dataframe
house_price_df = pd.DataFrame(house_price_datset.data, columns = house_price_datset.feature_names)
```

```python
house_price_df.head()
```

```python
# Add the Target Column to the dataFrame
house_price_df['House_Price'] = house_price_datset.target
```

```python
house_price_df.head()
```

```python
# Cheaking the row and columns in the dataset set
house_price_df.shape
```

```python
# cheak for missing valuse
house_price_df.isnull().sum()
```

```python
# Statstical Measures of the datset
house_price_df.describe()
```

# Finding the correlation betweem verious features in the dataset
1] Positive correlation

2] Negative correlation

```python
correlation = house_price_df.corr()
```

```python
#constructiong a heatmap to understand the correlation
plt.figure(figsize = (10,10))
sns.heatmap(correlation, cbar = True , square=True, fmt = '.1f', annot = True, annot_kws = {'size':8}, cmap = 'Blues')
```

# Splitting the Data and the Target


```python
X = house_price_df.drop(['House_Price'], axis = 1)
Y = house_price_df['House_Price']
```

```python
print(X)
```

```python
print(Y)
```

#Splitting the data into Trainig data and Test data

```python
X_train , X_test , Y_train , Y_test =train_test_split(X ,Y , test_size=0.2 ,random_state = 2)
```

```python
print(X.shape , X_train.shape ,X_test.shape )
```

# Model Training

XGBoost Regressor Model (Dicison Tree Algorithem)

```python
#loading the model
model = XGBRegressor()
```

```python
# Training (fitting) the model with X_train
model.fit(X_train , Y_train)
```

# Evaluation

prediction on training data

```python
#accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train , training_data_prediction)

# Mean Absolute ERROR
score_2 = metrics.mean_absolute_error(Y_train , training_data_prediction)

print("R squared error : " , score_1)
print("Mean Absolute ERROR : " , score_2)
```

Visualizing the Actual price and pridicted price

```python
plt.scatter(Y_train , training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show
```

prediction on test data

```python
#accuracy for prediction on training data
testing_data_prediction = model.predict(X_test)
```

```python
print(testing_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test , testing_data_prediction)

# Mean Absolute ERROR
score_2 = metrics.mean_absolute_error(Y_test , testing_data_prediction)

print("R squared error : " , score_1)
print("Mean Absolute ERROR : " , score_2)
```

```python
plt.scatter(Y_test , testing_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show
```
