Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Data Collection and processing

```python
# loding the datafrom cvs file to pandas  DataFrame
calories = pd.read_csv('/content/calories.csv')
```

```python
calories.head()
```

```python
# loding the datafrom cvs file to pandas  DataFrame
exercise = pd.read_csv('/content/exercise.csv')
```

```python
exercise.head()
```

Combining the two dataFrame

```python
calories_data = pd.concat([exercise , calories['Calories']] ,axis=1)
```

```python
calories_data.head()
```

```python
# cheaking the number of rows and columns
calories_data.shape
```

```python
# getting some info anout dataset
calories_data.info()
```

```python
# cheaking for missing values
calories_data.isnull().sum()
```

Data Analysis

Converting the test data to numerical Values

```python
calories_data.replace({'Gender':{'male':0,'female':1}}, inplace=True)
```

```python
calories_data.head()
```

Seprating the Feature and the Targets

```python
X = calories_data.drop(columns=['User_ID','Calories'], axis=1)
Y = calories_data['Calories']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(calories_data.shape , X_train.shape , X_test.shape)
```

Training the Model

XGBoost Regressor

```python
# Loding the model
regressor = XGBRegressor()
```

```python
# Training the model
regressor.fit(X_train, Y_train)
```

Evaluation

Prediction on test data

```python
test_data_prediction = regressor.predict(X_test)
```

```python
print(test_data_prediction)
```

Mean absolute error

```python
# finding Mean absolute error
mae = metrics.mean_absolute_error(Y_test, test_data_prediction)
```

```python
print("Mean Absolute Error : " ,mae)
```
